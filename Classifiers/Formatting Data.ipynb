```python
import pandas as pd
import numpy as np
```

```python
bankrupt_df = pd.read_csv("Bankrupt Companies Bloomberg DB.csv")
financial_df = pd.read_csv("Financials All Listed.csv")
identity_df = pd.read_csv("Identity.csv")
```

```python
bankrupt_df = bankrupt_df[['Ticker','Security','Revenue','Mst Rec Bnkrpty Inslvency Dt']]
bankrupt_df = bankrupt_df.dropna()
bankrupt_df.rename(columns={'Mst Rec Bnkrpty Inslvency Dt':'Insolvency Date'},inplace = True)
bankrupt_df.rename(columns={'Security':'Name'},inplace=True)
bankrupt_df
```

```python
# code to remove the " IN" portion of the Ticker
bankrupt_df['Ticker'] = bankrupt_df['Ticker'].str.replace(' IN', '', regex=False)
bankrupt_df
```

```python
bankrupt_df = bankrupt_df.drop_duplicates()
```

```python
bankrupt_df = bankrupt_df.drop(columns = ['index','level_0'])
bankrupt_df
```

```python
bankrupt_df['Name'] = bankrupt_df['Name'].str.upper()
```

```python
bankrupt_df['Name'] = bankrupt_df['Name'] + '.'
```

```python
bankrupt_df
```

```python
bankrupt_df.to_csv("bankrupt_final.csv")
```

```python
merged_data = financial_df.merge(bankrupt_df,how='left',on='Name')
merged_data
```

```python
merged_data.head(1000)
```

```python
merged_data.to_csv('merged_data.csv')
```

```python
count_bankrupt_matched = merged_data['Revenue'].notna().sum()
count_bankrupt_matched
```

```python
merged_data['Label'] = merged_data['Insolvency Date'].notna().astype(int)
merged_data.head()
```

```python
merged_data.to_csv('merged_data.csv')
```
